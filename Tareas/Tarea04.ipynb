## <b>Recrea la siguiente figura</b>

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generar datos para la distribuci√≥n normal
x = np.linspace(-10, 10, 50)
mu = 0
sigma1 = 3.3
sigma2 = 4.2
y1 = 1.1 * np.exp(-(x - mu)**2 / (2 * sigma1**2))
y2 = np.exp(-(x - mu)**2 / (2 * sigma2**2))

# Crear el gr√°fico principal sin grid
sns.set_style("white")
fig, ax = plt.subplots(figsize=(8, 6))
ax.grid(False)

# Distribuci√≥n normal m√°s aplanada (l√≠nea punteada con puntos y l√≠neas m√°s grandes)
ax.plot(x, y1, color='red', linewidth=3, linestyle='dashdot', markersize=7, markerfacecolor='red', markeredgecolor='black')

# Distribuci√≥n normal m√°s amplia (l√≠nea s√≥lida con puntos amarillos y l√≠neas delgadas m√°s grandes)
ax.plot(x, y2, color='blue', linewidth=1, marker='o', markersize=9, markerfacecolor='gold', markeredgecolor='black', markeredgewidth=2)

# Configuraci√≥n adicional del gr√°fico principal
ax.set_title('Many Plots', fontsize=16)

# Llamar eje x That took a while
ax.set_xlabel('That took a while', fontsize=18)


# Leyendas en formato LaTeX
ax.legend([r'$3e^{-0.25x^2}$', r'$2.8e^{-0.15(x-0.1)^2}$'], fontsize=12)

# Agregar las palabras "cool", "plot" y "bro" en los colores y ubicaciones especificadas
ax.text(-2, 0.5, 'Cool', color='red', fontsize=20, ha='center', va='center')
ax.text(0, 0.4, 'Plot', color='blue', fontsize=20, ha='center', va='center')
ax.text(2, 0.3, 'Bro', color='green', fontsize=20, ha='center', va='center')

# Mostrar el gr√°fico
plt.show()
```

## <b> Recrea el siguiente  `violinplot` comparando la distribuci√≥n de  `Entire home/apt` y `Private room` para los 5 `neighbourhood groups`</b>

```python
#Ingresa tu c√≥digo aqu√≠


df = abnb

# Establecer el estilo de los gr√°ficos de Seaborn
sns.set(style='whitegrid')

# Filtrar el DataFrame para tener solo dos niveles para room_type
df_filtered = df[df['room_type'].isin(['Entire home/apt', 'Private room'])]

# Crear el violinplot
sns.violinplot(data=df_filtered, x='neighbourhood_group', y='price', hue='room_type', split=True)

# Ajustar la leyenda
plt.legend(title='Room Type')

# Mostrar el gr√°fico
plt.show()
```


## <b>Time Series: Recrea el siguiente gr√°fico</b>

(Pista: Convierte la columna `last_review` a formato `DateTime`)

```python
#Ingresa tu c√≥digo aqu√≠

df = abnb
# Convertir la columna 'last_review' al tipo de datos datetime
df['last_review'] = pd.to_datetime(df['last_review'])
# Dejar datos desde el 2016-10 hasta el 2019-04
df = df[(df['last_review'] >= '2016-10') & (df['last_review'] <= '2019-04')]

# Ordenar el DataFrame por la columna 'last_review'
df = df.sort_values('last_review')

# Crear la gr√°fica de serie de tiempo
plt.figure(figsize=(10, 6))

plt.plot(df['last_review'], df['number_of_reviews'], label='Number of Reviews')
plt.plot(df['last_review'], df['price'], label='Price')

# Establecer los t√≠tulos y etiquetas de los ejes
plt.title('Fluctuations in Number of Reviews and Price over time')
# nombrar el eje x
plt.xlabel('last_review')

# A√±adir una leyenda
plt.legend()

# Mostrar la gr√°fica
plt.show()
```
