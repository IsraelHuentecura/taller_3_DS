 # Pandas 04 **Combinando DataFrames con Pandas**

> ## Distribuciones
>
> Cree un nuevo DataFrame uniendo los contenidos de survey.csv y species.csv. Luego calcula:
>
>   - taxa por parcela
>   - taxa por sexo por parcela


```python
surveys_df = pd.read_csv('./pandas_data/pandas_data/surveys.csv')
species_df = pd.read_csv('./pandas_data/pandas_data/species.csv')
```

```python


# unir los DataFrames usando la columna 'species_id'
merged = pd.merge(surveys_df, species_df, on='species_id')

# calcular taxa por parcela
taxa_por_parcela = merged.groupby('plot_id')['taxa'].nunique()

taxa_por_parcela
```

```python
# calcular taxa por sexo por parcela
taxa_por_sexo_por_parcela = pd.pivot_table(merged, index='plot_id', columns='sex', values='taxa', aggfunc=len)
taxa_por_sexo_por_parcela
```

> ## Índice de Diversidad
> - En la carpeta de datos, hay un archivo plots.csv que contiene información sobre el tipo asociado con cada parcela. Usa esos datos para resumir el número de parcelas por tipo de parcela.
>  - Calcula un índice de diversidad de su elección para control vs parcelas de rodamiento de roedores. El índice debe considerar tanto la abundancia de especies como el número de especies. Puedes optar por utilizar el simple índice de biodiversidad descrito aquí que calcula la diversidad como:
> el número de especies en la parcela / el número total de individuos en la parcela = índice de biodiversidad.

```python
plots_df = pd.read_csv("./pandas_data/pandas_data/plots.csv ")

# Agrupar los datos por tipo de parcela
grupos = plots_df.groupby("plot_type").sum()
grupos
```

```python
# filtrar para incluir solo los datos para control y parcelas de rodamiento de roedores
parcelas_control = plots_df[plots_df.plot_type == 'Control']['plot_id']
parcelas_rodamiento = plots_df[plots_df.plot_type == 'Rodent Exclosure']['plot_id']
merged_filtrado = merged[(merged.plot_id.isin(parcelas_control)) | (merged.plot_id.isin(parcelas_rodamiento))]
```

```python
# calcular el número total de individuos y el número de especies en cada parcela
parcelas_info = merged_filtrado.groupby(['plot_id', 'plot_type'])['species_id'].agg([('n_species', 'nunique'), ('n_individuos', 'count')])
parcelas_info
```

```python
# calcular el índice de biodiversidad para cada parcela
parcelas_info['biodiversidad'] = parcelas_info.apply(lambda row: row['n_species'] / row['n_individuos'], axis=1)
parcelas_info
```
