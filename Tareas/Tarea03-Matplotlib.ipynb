```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('./data/for007.dat')

# Grfico 1
plt.subplot(4, 2, 1) # 2 filas, 4 columnas, posicin 1
plt.plot(data[:,1], data[:,2], 'b')
plt.xlabel("Col|2")
plt.ylabel("Col|3")
plt.legend(["HYWIND"])
plt.title("Velocity profile")
plt.grid(linestyle='dashed')

# Grfico 2
plt.subplot(4, 2, 2) # 2 filas, 4 columnas, posicin 5
plt.plot(data[:30,0] , data[:30,2], 'b-')
plt.xlabel("Col|1")
plt.ylabel("Col|3")
plt.title("V(r/R)(inner)")
plt.ylim(0,20)
plt.grid(linestyle='dashed')


# Grfico 3
plt.subplot(4, 2, 3) # 2 filas, 4 columnas, posicin 2
plt.plot(data[:,1] , data[:,2], 'b-')
plt.plot(data[:,1] , data[:,16], 'r-')
plt.xlabel("Col|2")
plt.ylabel("Col|3(blue); Col|17(red)")
plt.legend(["HYWIND", 'TRIAL PROFILE'])
plt.title("Velocity profile")
plt.grid(linestyle='dashed')


# Grfico 4
plt.subplot(4, 2, 4) # 2 filas, 4 columnas, posicin 6
plt.plot(data[:,0] , data[:,2], 'b-')
plt.xlabel("Col|1")
plt.ylabel("Col|3")
plt.title("V(r/R)(outer)")
plt.ylim(380,420)
plt.xlim(70,100)
plt.grid(linestyle='dashed')


# Grfico 5
plt.subplot(4, 2, 5) # 2 filas, 4 columnas, posicin 3
plt.plot(data[:,1] , data[:,3], 'b-')
plt.xlabel("Col|2")
plt.ylabel("Col|4")
plt.title("dw/du")
plt.grid(linestyle='dashed')


# Grfico 6
plt.subplot(4, 2, 6) # 2 filas, 4 columnas, posicin 7
plt.plot(np.log10(data[:,0]-1.0) , np.log10(data[:,5]), 'b-')
plt.xlabel("log (col|1-1.0)")
plt.ylabel("log (col|6)")  
plt.title("Density")
plt.grid(linestyle='dashed')


# Grfico 7
plt.subplot(4, 2, 7) # 2 filas, 4 columnas, posicin 4
plt.plot(data[:880,0] , data[:880,8], 'b-')
plt.xlabel("Col|2")
plt.ylabel("Col|4")
plt.title("g_line")
plt.grid(linestyle='dashed')


# Ajustar el tamao y la separacin entre los grficos
plt.subplots_adjust(left=0.05, bottom=0.1, right=0.95, top=0.9, wspace=0.7, hspace=0.5)
# Ajustar el tamao de la figura
plt.gcf().set_size_inches(10, 14)

# Mostrar los grficos
plt.show()
```
